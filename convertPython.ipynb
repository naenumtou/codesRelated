```python
# Import library
import pandas as pd
import numpy as np
import re
import m2cgen as m2c #Converter

from xgboost import XGBRegressor
```

```python
# Import data
df = pd.read_csv('https://raw.githubusercontent.com/naenumtou/statisticalModel/main/datasets/Bayesian_model.csv')

# Target and features split
X = df.drop(['y'], axis = 1)
y = df['y']

# Training model
model = XGBRegressor()
model.fit(X, y)
```

```python
# Model prediction
pred = pd.DataFrame(model.predict(X), columns = ['Prediction'])
pred.head(10)
```

```python
# Covert code to VBA
codeVBA = m2c.export_to_visual_basic(model, function_name = 'pred')
```

```python
# Print VBA Code
print(codeVBA)
```

```python
# Convert to SAS
# Remove VBA Syntax
code = re.sub('Dim var.* As Double', '', codeVBA)
code = re.sub('End If', '', code)

# Add ';' for SAS Syntax
# After values
valuesList = re.findall('[0-9]+\n', code)

for i in range(len(valuesList)):
  foundString = valuesList[i]
  newString = valuesList[i][:-1] + ';\n' #Insert add the end to string
  code = code.replace(foundString, newString)

# After string
stringsList = re.findall('\)\n', code) # For prediction line

for i in range(len(stringsList)):
  foundString = stringsList[i]
  newString = stringsList[i][:-1] + ';\n' #Insert add the end to string
  code = code.replace(foundString, newString)

# Change 'inputVetor(x)' to variable name
# Create dictionary for key mapping
inputVec = {'inputVector(0)':'x1',
            'inputVector(1)':'x2',
            'inputVector(2)':'x3'} 
            
for varName in inputVec.keys():
  code = code.replace(varName, inputVec[varName])


# Change input to dataset step
code = re.sub('Module Model\n', 
              'DATA pred;\n', # DATA;
              code)

code = re.sub('Function pred\(ByRef inputVector\(\) As Double\) As Double\n', 
              'SET input_data;\n', #SET;
              code)

code = re.sub('End Function\nEnd Module\n',
              'RUN;', #RUN;
              code)

print(code)
```

```python
# Export SAS Code
sasCode = open('XGBoostModel.sas', 'w')
sasCode.write(code)
sasCode.close()
```
